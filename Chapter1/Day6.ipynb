###  Inheritance

```python
#base class
class Animal :
    def __init__(self,name):
        self.name = name
        
    def speak(self):
        print(f"{self.name} is speaking")  
        
        
#derived class

class dog(Animal):
    pass

d1 = dog("Lucky")
d1.speak()         
```

```python
#Simple inheritance
#base class
class Animal :
    def __init__(self,name):
        self.name = name
        
    def speak(self):
        print(f"{self.name} is speaking")  
        
        
#derived class

class dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed
        
    def speak(self):
        super().speak()
        print(f" {self.breed} {self.name} is barking")    

d1 = dog("Lucky", "Shibu Inu")
d1.speak()         
```

```python
#Multiple inheritance

class A:
    def show(self):
        print("A")
        
class B:
    def show(self):
        print("B")    
        
class C(A, B):
    pass

c1 = C()
c1.show ()           
```

```python
# Method Resolution Order
C.mro()
```

```python
class DataList:
    def __init__(self, alist):
        self.alist = alist
       
        
    def mean(self):
        return sum(self.alist)/len(self.alist)   
  
```

```python
l1 = DataList([1,2,3,4,5])

l1.mean()
```

```python
while True:
    try:
        n1 = input("First Number (or 'q' to quit): ")
        if n1.lower() == 'q':
            print("Exiting program...")
            break

        n2 = input("Second Number (or 'q' to quit): ")
        if n2.lower() == 'q':
            print("Exiting program...")
            break

        # Convert to integers after checking for quit
        n1 = int(n1)
        n2 = int(n2)

        result = n1 / n2
        print(f"Result is {result}")

    except ZeroDivisionError:
        print("Cannot divide by Zero")

    except ValueError:
        print("Please enter valid number")

    finally:
        print("This always runs")
```
