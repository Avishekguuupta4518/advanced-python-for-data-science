### WAP to find the factorial of a number


```python
n = 5
f = 1

while n >= 1:
    f *= n
    n -= 1

print(f)
```

### Write a function to find the sum of first n natural numbers

```python
def sum(n):
    s = 0
    for i in range(1, n+1):
        s += i    
    return s

sum(10)
```

### calculate average using *args

```python
def average(*args):
    s = 0
    for num in args:
        s += num
    return s/len(args)    
        
average(1, 2, 3,4,5,6,7)
```

### use of **kwargs in a function

```python
def myFunction(**kwargs):
    print(kwargs)
    print(type(kwargs))
    
myFunction(a=10, b=20)
```

### to find the sum of values using **kwargs

```python
def myFunction(**kwargs):
    s = 0
    for key in kwargs.keys():
        s += kwargs[key]
    return s
myFunction(a=10, b=20)  
```

### OOP concept class and objects

```python
class Animal:
    def __init__(self, name):
        self.name = name
        
    def __str__(self):
        return f'Animal name is {self.name}'  
    
    def __repr__(self):
        return f'Animal : {self.name}'   
        
    def sayHello(self):
        print(f'{self.name} says HEllo !')    
        
a1 = Animal("Tommy")
print(a1)
a1
a1.sayHello() 
```

### operator overloading in a class

```python
class Point:
    def __init__(self,xx,yy):
        self.x = xx
        self.y = yy 

    def __str__ (self):
        return f'({self.x}, {self.y})'
    
    def __add__(self,other):
        new_x = self.x + other.x
        new_y = self.y + other.y
        return Point(new_x, new_y)
    
p1 = Point(1,2)
p2 = Point(2,1)
p3 = p1+p2 # p1.__add__(p2)
print(p3) 
```

### WAP to create a class named Distance which has feet and inches as data member. Overload any two operators

```python
class Distance:
    def __init__(self, feet=0, inch=0):
        self.feet = feet
        self.inch = inch
        self.normalize()
    
    def normalize(self):
        if self.inch >=12:
            self.feet += self.inch // 12
            self.inch = self.inch % 12
              
    def __str__(self):
        return f'{self.feet} feet {self.inch} inch'
    
    def __add__(self,other):
        f = self.feet + other.feet
        i = self.inch + other.inch
        return Distance(f, i)
    
d1 = Distance(5, 8)
d2 = Distance(3, 9)
d3 = d1 + d2 
print("total distance :" , d3)    
```
