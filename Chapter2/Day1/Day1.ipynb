```python
with open(file="try.txt") as f:
    content = f.read()
    print(len(content.split()))
    print(content)
```

```python
with open(file="try.txt") as f:
    content = f.readline()
    for line in content:
        print(line)
```

```python
with open(file="demo.txt", mode="w") as f:
    while(True):
        sentence = input("Enter a sentence (Press Q or q to quit)")
        if sentence == 'Q' or sentence == 'q':
            break
        f.writelines(sentence)
```
