```python
import csv

data = [
    ['Name', 'Address', 'Age'],
    ['Ram', 'Dharan', 22],
    ['Shyam', 'Brt', 24],
    ['Sita', 'Ktm', 21]
]

# Writing to CSV file
with open("newcsv.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)

print("CSV file created successfully!")
```

```python
import csv

data = [
    {'Name': 'Ram', 'Address': 'Dharan', 'Age': 22},
    {'Name': 'Shyam', 'Address': 'Brt', 'Age': 23},
    {'Name': 'Sita', 'Address': 'Ktm', 'Age': 24},
]

with open("newnew.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["Name", "Address", "Age"])
    writer.writeheader()       # writes the header row
    writer.writerows(data)     # writes all dictionary rows

print("CSV file created successfully!")
```

```python
import pandas as pd

# Create DataFrame with student marks
new_df = pd.DataFrame([
    {'Roll': 1, 'Name': 'Ram', 'Eng': 80, 'Sci': 97, 'Math': 94},
    {'Roll': 2, 'Name': 'Shyam', 'Eng': 75, 'Sci': 88, 'Math': 90},
    {'Roll': 3, 'Name': 'Sita', 'Eng': 85, 'Sci': 92, 'Math': 89}
])

new_df['Total'] = new_df[['Eng', 'Sci', 'Math']].sum(axis=1)
new_df['Percentage'] = (new_df['Total'] / 3).round(2)

print(new_df)
```
