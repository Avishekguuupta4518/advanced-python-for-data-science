```python
import pandas as pd

data = {
    "Name": ["Amit", "Riya", "Suman", "Karan", "Neha"],
    "Math": [78, 55, 90, 45, 68],
    "Science": [65, 60, 88, 50, 72],
    "English": [70, 58, 92, 40, 75]
}

df = pd.DataFrame(data)
df.to_excel("marks.xlsx", index=False)

df["Total"] = df["Math"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3

df["Result"] = df["Average"].apply(lambda x: "Passed" if x >= 60 else "Failed")

df.to_excel("students_result.xlsx", index=False)

print("✅ students_result.xlsx generated successfully.")
```

```python
import pandas as pd

df = pd.read_excel("marks.xlsx", engine="openpyxl")

df["Total"] = df["Math"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3

df["Average"] = df["Average"].round(2)

df["Result"] = df["Average"].apply(lambda x: "Passed" if x >= 60 else "Failed")

df.to_excel("students_result.xlsx", index=False, engine="openpyxl")

print("✅ students_results.xlsx has been created successfully!")
```
